```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
x = np.array([4.0,4.2,4.5,4.7,5.1,5.5,5.9,6.3,6.8,7.1])
y = np.array([102.56,113.18,130.11,142.05,167.53,195.14,224.87,256.73,299.50,326.72])
```

```python
def exp_function(x, alpha, beta):
    return alpha * np.exp(beta * x)

params, covar = curve_fit(exp_function, x, y)

alpha, beta = params

y_fit = exp_function(x, alpha, beta)
```

```python
y_log = np.log(y)

def linear_function(x, a0, a1):
    return a0 + a1*x

params, covar = curve_fit(linear_function, x, y_log)

a0, a1 = params

y_fit = exp_function(x, a0, a1)
```

```python
x_fit_lin = np.linspace(4.0, 7.1, 100)
y_fit_lin = np.exp(linear_function(x_fit_lin, a0, a1))

x_fit_exp = np.linspace(4, 7.1, 100)
y_fit_exp = exp_function(x_fit_exp, alpha, beta)

plt.scatter(x, y, label='Data')
plt.plot(x_fit_lin, y_fit_lin, 'r', label='Best Fit (Linear)')
plt.plot(x_fit_exp, y_fit_exp, 'g', label='Best Fit (Exponential)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
